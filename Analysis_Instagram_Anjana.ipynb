```python
import os
import pandas as pd

def main_formula(directory): 
    # Initialize an empty list to store DataFrames
    dataframes = []
    
    # Loop through each file in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):  # Check if the file is a CSV file
            filepath = os.path.join(directory, filename)
                # Read the CSV file into a DataFrame
            df = pd.read_csv(filepath)
                # Append the DataFrame to the list
            dataframes.append(df)
                
    # Step 2: Compare the column names and order
    all_columns = set(dataframes[0].columns)
    for df in dataframes[1:]:
        all_columns = all_columns.union(set(df.columns))
    
    # Step 3: Reorganize the columns
    for df in dataframes:
        # Add missing columns as placeholders
        missing_columns = list(all_columns - set(df.columns))
        for column in missing_columns:
            df[column] = None
        # Reorder columns
        df = df.reindex(columns=all_columns)
    
    # Step 4: Concatenate all the DataFrames
    master_df = pd.concat(dataframes, ignore_index=True)
    return master_df

```

```python
import re

# Load your Instagram data into a pandas DataFrame
instagram_data = master_df

# Define a function to extract hashtags from text
def extract_hashtags(text):
    hashtags = re.findall(r"#(\w+)", text)
    return hashtags

# Apply the function to the "Description" column to extract hashtags
instagram_data['Hashtags'] = instagram_data['Descripción'].apply(extract_hashtags)

# Create a list of all hashtags
all_hashtags = [hashtag for sublist in instagram_data['Hashtags'] for hashtag in sublist]

# Count the occurrences of each hashtag
hashtag_counts = pd.Series(all_hashtags).value_counts()

# Display the top 10 hashtags
top_10_hashtags = hashtag_counts.head(10)
print(top_10_hashtags)
```
